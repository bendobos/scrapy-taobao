```python
td = (datetime.now() - timedelta(0)).strftime('%Y%m%d%H')
yd = (datetime.now() - timedelta(1)).strftime('%Y%m%d%H')
```

```python
# df_td = read_csv('taobao.%s.csv' % td)
# df_yd = read_csv('taobao.%s.csv' % yd)
df_td = read_csv('taobao.morel.%s.csv' % '2016101612')
df_yd = read_csv('taobao.morel.%s.csv' % '2016101520')
```

```python
df_yd.shape
```

```python
df_td['id'] = df_td['id'].astype(str).astype('category')
df_yd['id'] = df_yd['id'].astype(str).astype('category')

df_td['view_sales'] = df_td['view_sales'].astype(float)
df_yd['view_sales'] = df_yd['view_sales'].astype(float)
```

```python
__vars = ['id', 'query', 'raw_title']
_vars = __vars + ['view_sales']
df = merge(
    df_td.ix[:, _vars],
    df_yd.ix[:, _vars],
    on=__vars,
    how='outer',
    suffixes=['_td', '_yd']
)
df = merge(
    df,
    df_td.ix[:,['id','rank','view_price', 'is_tmall', 'item_loc', 'is_ad', 'sold_total_count', 'category', 'sc_service_rating']],
    on='id',
    how='left',
)
```

```python
df['daily_sales'] = df['view_sales_td'] - df['view_sales_yd']
```

```python
df['is_soup'] = df['raw_title'].str.contains('汤')
df['is_wild'] = df['raw_title'].str.contains('野生')
```

```python
_df = df_td[df_td['category'] == 50009778]
_df['sales_amount_30d'] = _df['sold_total_count'] * _df['view_price']
_df['is_wild'] = _df.raw_title.str.contains('野生')
__df = DataFrame(_df.groupby('is_wild')['sales_amount_30d'].sum())
__df2 = __df / _df['sales_amount_30d'].sum()
__df2.columns = ['pct']
__df.join(__df2)
```

```python
# 野生羊肚菌
# __df = _df[_df.is_wild]
__df = _df.copy()
__df = __df.sort_values('rank')[['id', 'raw_title', 'is_wild', 'rank', 'sales_amount_30d', 'reserve_price',
                                 'view_sales', 'sold_total_count', 'view_price', 'unit_price']]

mask = (__df['rank'] < 200) & (__df['view_price'] < 500)
x, y = __df[mask]['view_price'], __df[mask]['sold_total_count']
print 'wild morel'
sns.jointplot(x, y, xlim=(0, 500))
plt.show()
g = sns.FacetGrid(__df[mask], hue='is_wild', size=6)
g.map(plt.scatter, 'view_price', 'sold_total_count', alpha=0.5)
g.add_legend()
```
